#**Caso De Estudio: Detección de Riesgo SARLAFT en una Entidad Financiera**

En este documento se desarrolla y analiza un modelo Naive Bayes para la detección de riesgo SARLAFT (lavado de activos) en una entidad financiera utilizando diferentes variables socioeconómicas y basándose en el comportamiento financiero y los  cambios inusuales en su patrimonio; para de esta manera determinar si las personas tienen un alto riesgo de incurrir en lavado de activos o no lo tienen.

Esto se realiza debido a que la entidad a pesar de contar con unos controles tradicionales de monitoreo, detectó que se presentaron operaciones sospechosas que no se identificaron de manera oportuna.


De aucerdo con lo anterior, las variables tomadas en cuenta se explican a continuación:

**Edad:** es la edad de la persona en el momento en que se está realizando su evaluación para saber si incurre en lavado de actvivos o no.

**Ingresos:** son los ingresos recibidos por una persona como resultado de una actividad comercial.

**Egresos:** son los gastos que tiene cada una de las personas (se tienen en cuenta gastos de todo tipo).

**Monto Transado:** es el monto que las personas movieron en sus tarjetas de crédito durante el periodo de tiempo que se realizó el seguimiento a las transacciones.

**Número de Tarjetas de Crédito:** cantidad de tarjetas de crédito que posee la persona al momento de la evaluación.

**Porcentaje Crecimiento Patrimonio:** es el crecimiento del patrimonio expresado en porcentaje, para analizar cuánto aumentó o disminuyó con respecto a la anterior información. Entendiendo el patrimonio como el conjunto total de sus bienes (materiales e inmateriales) valorables económicamente al momento de la evaluación.

Cabe aclarar que todas las variables están expresadas en USD, exceptuando la edad, el porcentaje de crecimiento del patrimonio y el número de tarjetas de crédito.


***0. Se procede con la carga de las librerías de trabajo***



```python
import numpy as np
import pandas as pd

#Librerías específicas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix

from google.colab import drive
drive.mount('/content/drive')
```

***1. Se cargan los datos de trabajo de la base de datos LavadoActivos***

```python
nxl = '/content/drive/MyDrive/Analítica de Negocios/2. LavadoActivos.xlsx'
XDB = pd.read_excel(nxl,sheet_name=0)
XDB.head(100)

#Seleccionamos variables de trabajo
XD = XDB [['edad','ingresos','gastos','monto transado', 'tarjetas credito', 'porcentaje_crecimiento_patrimonio']]
XD.head(10)
yd = XDB[['lavado_activos']] #Variable de decisión
yd.head()
```

***2. Se implementa el Modelo Naive Bayes***

```python
np.set_printoptions(suppress=True,precision=3) #3 decimales
mnb = GaussianNB()
mnb.fit(XD,yd) #Ajustar variables Entrada - Salida

#Mostramos las medidas de las variables
u = mnb.theta_
sigma = mnb.var_ ; sigma=np.sqrt(sigma)
print("'edad','ingresos','gastos','monto transado', 'tarjetas credito', porcentaje_crecimiento_patrimonio'")
print(u)
print()
print()
print("Las desviaciones son: ")
print(sigma)
print()
print()
print("Los límites superiores de las variables son: ",u+sigma)
print()
print()
print("Los límites inferiores de las variables son: ",u-sigma)
print()
print()
```

De igual manera, se puede evidenciar que si bien la media de las personas que incurren en lavado de activos tienen una media de 44 y esta puede subir hasta 59 y bajar hasta 29 (es decir, oscila entre esas edades); y para las personas que no incurren en lavado de activos, tienen una media de 45 y puede subir hasta 60 y bajar hasta 30 (es decir, oscila entre esas edades).

Por otro lado, con respecto a los ingresos y a los gastos de estas personas, se puede evidenciar que para el caso de ambos, las personas que no incurren en lavado de activos tienen tanto ingresos como gastos menores que las personas que si incurren en lavado de activos; sin embargo la presencia simultánea de ingresos y gastos elevados no constituye por sí sola una conducta ilícita, pero si se combina con otros factores que se evalúan en el modelo, esto se convierte en un elemento importante para identificar alertas de SARLAFT. Por ejemplo, si se combina con el monto transado (que también fue una variable que se analizó), se evidencia que este también es mayor en las personas que son opcionadas para incurrir en lavado de activos que en las que no son opcionadas, y esto combinado con los altos ingresos y gastos, sugiere que hay una alta rotación y volumen de capital (en este caso el monto transado de las personas que incurren en lavado de activos es de 33835 aprox y de las personas que no incurren de 3090 aprox).

Se puede observar que tanto el número promedio de tarjetas de crédito como el porcentaje de crecimiento del patrimonio presentan valores muy similares entre los clientes que son opcionados y no opcionados en incurrir en lavado de activos; y se puede deducir que estas variables no actúan como factores determinantes en la identificación del riesgo de lavado de activos, dándonos una idea de que el riesgo se manifiesta principalmente en la dinámica y volumen de las transacciones, más que en la acumulación patrimonial o en el acceso que tengan a productos financieros; es decir, el riesgo no está en cuánto patrimonio tienen ni en cuántas tarjetas de crédito tienen, sino en cómo y en cuánto dinero se mueve.
